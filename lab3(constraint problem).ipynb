-----Constraint Satisfaction Problem-----


```python
!pip install python-constraint
from constraint import Problem
scheduler = Problem()

courses = ['Algebra101', 'Chemistry101']
time_slots = ['Tue9', 'Tue10']
rooms = ['HallA', 'LabB']

room_types = {
    'HallA': 'Lecture',
    'LabB': 'Lab'
}

course_requirements = {
    'Algebra101': 'Lecture',
    'Chemistry101': 'Lab'
}

valid_domains = {}
for course in courses:
    valid_domains[course] = [
        (slot, room) for slot in time_slots
        for room in rooms
        if room_types[room] == course_requirements[course]
    ]
    scheduler.addVariable(course, valid_domains[course])

def teacher_conflict(x, y):
    return x[0] != y[0]

scheduler.addConstraint(teacher_conflict, ['Algebra101', 'Chemistry101'])

def room_conflict(x, y):
    return not (x[0] == y[0] and x[1] == y[1])

scheduler.addConstraint(room_conflict, ['Algebra101', 'Chemistry101'])

solutions = scheduler.getSolutions()

for i, sol in enumerate(solutions, start=1):
    print(f"Solution {i}:")
    for course in courses:
        print(f"  {course}: Time = {sol[course][0]}, Room = {sol[course][1]}")
    print("-" * 30)
```
